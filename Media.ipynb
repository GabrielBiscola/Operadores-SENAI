```python
print("Vamos calcular a media do aluno")
n1 = int(input("Entre com a primeira nota:"))
n2 = int(input("Entre com a segunda nota:"))
media = (n1+n2)/2

if media >= 7:
  print("Aluno aprovado")
else:
  print("Deu ruim. Voce ficou {} pontos abaixo da media.".format(7-media))
```
